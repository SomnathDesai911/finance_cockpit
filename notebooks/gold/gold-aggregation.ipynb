```python
df_silver=read_delta_table(spark,"cockpit.finance.silver_table")
display(df_silver)
```

```python
df_silver=(
    df_silver
    .withColumn("SLA_Breached",when(col("Hours")>30,"Yes").otherwise("No"))
)
df_silver_breached_sla=(
    df_silver
    .groupBy("SLA_Breached").agg(count("*").alias("count"))      
    )
df_silver_breached_sla.show()SLA_Breached
```
